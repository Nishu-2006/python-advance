```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
#write a program to input initial velocity and acceleration and plot the graph depicting the equation.
#v=u+a*t
#s=u*t+0.5a*t*2
#v*2=u*2+2*a*s
u=float(input("enter initial velocity"))
a=float(input("enter initial acceleration"))
t=np.linspace(0,10,100)
v=u+a*t
s1=u*t+0.5*a*t**2
s2=(v**2-u**2)/2*a
plt.plot(v,t,color="red")
plt.plot(s1,v,color="green")
plt.plot(s1,t,color="pink")
plt.xlabel('time or velocity')
plt.ylabel('velocity/displacement')
plt.title('equation of motion')
plt.show()
```

```python
fig,ax=plt.subplots()
fruits=['apple','banana','cherry']
counts=[40,60,80]
bar_labels=['red','yellow','blue']
bar_colors=['tab:red','tab:orange','tab:blue']
ax.bar(fruits,counts,label=bar_labels,color=bar_colors)
ax.set_ylabel('fruit supply')
ax.set_title('fruit supply by kind and color')
ax.legend(title='fruit color')
plt.show()
```

```python
species=('adelie','chinstrap','gentoo')
sex_count={'Male':np.array([73,34,61]),'female':np.array([73,34,58])}
width=0.6
fig,ax=plt.subplots()
bottom=np.zeros(3)
for sex,sex_count in sex_count.items():
    p=ax.bar(species,sex_count,width,label=sex,bottom=bottom)
    bottom =sex_count
    ax.bar_label(p,label_type="center")
ax.set_title("Pengins by sea")
ax.legend()
plt.show()
```

```python
#stacked plot
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
sales_car=[120,30,56,89,90,120,34,56,78,12,45,56]
sales_bike=[240,27,34,56,98,110,150,34,10,23,72,78]
plt.stackplot(months,sales_car,sales_bike,labels=['car','bike'])
plt.legend()
plt.title("car vs bike sale")
plt.show()
```
