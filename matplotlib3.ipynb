```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Loading a data set from online repository
a=sns.load_dataset('flights')
print(a)
```

```python
#to plot lines lineplot() is used:
sns.lineplot(x='year',y='passengers',data=a)
plt.show()
```

```python
#to plot lines lineplot() is used:
sns.lineplot(x='year',y='passengers',hue='month',data=a)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots: Dark Theme
plt.figure(figsize=(20,10))
sns.set(style='dark')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.barplot(x='month',y='passengers',data=a,palette='rocket')
plt.show()
```
