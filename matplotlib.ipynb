<font color='red'><h1>MATPLOTLIB:</h1></font>
- mathematical plotting library.
- used to visualize the data
- works on a 2D numpy array.
- used to draw graphs,bar chart,pie charts,scatter plot,maps,legend etc.


```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
#1. Line Plot
#plot x vs y line
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.plot(x,y)
#plot: plot is a function used to draw a 2D plot ar graph using values on x and y asis
plt.show()
```

<h2>#plot:</h2>
 plot is a function used to draw a 2D plot ar graph using values on x and y asis

```python
#plot x vs y line
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.plot(x,y,color='red')
plt.show()
```

```python
#addind title to the plot and label to axis
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x vs y plot")
plt.plot(x,y,color="pink")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
x1=[1,3,5,7,9,11,13]
y1=[5,6,7,8,9,10,11]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlabel("x1-axis")
plt.ylabel("y1-axis")
plt.title("x vs y plot")
plt.plot(x,y,color="pink")
plt.plot(x1,y1,color="green")
plt.show()
```

```python
#displaying legend in the plot
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.xlabel("x-asix")
plt.ylabel("y-axis")
plt.plot(x,y,color="red",linestyle="--",linewidth=5,label='Line-1')
a=[3,9,10,4,5]
b=[1,7,3,2,9]
plt.plot(a,b,color="blue",label="Line-2")
plt.legend(loc="best")
plt.show()
```

linestyle="solid"

linestyle="dashed"

```python
x=[1,8,3,13,6,5]
y=[2,4,21,12,3,5]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph with corner marking")
plt.plot(x,y,color="red",marker="o",markersize=10,markerfacecolor="blue",label="line1")
plt.show()

```

marking:----
'o'=circle
's'=square
'+'=plus
'D'=diamond
'^'=triangle
'*'=star
'X'=X
